```python
# Dealing with the null value 
```

```python
df2['match_winner'].fillna('Unknown', inplace=True)
df2['wb_runs'].fillna(0, inplace=True)
df2['wb_wickets'].fillna(0, inplace=True)
df2['player_of_the_match'].fillna('No Award', inplace=True)
df2['first_ings_score'].fillna(0, inplace=True)
df2['balls_left'].fillna(0, inplace=True)
df2['first_ings_score'].fillna(0, inplace=True)
df2['second_ings_score'].fillna(0, inplace=True)
df2['highscore'].fillna(0, inplace=True)
df2['toss_decision'].fillna(df['toss_decision'].mode()[0], inplace=True)
```

### 1. How many matches were played at each venue?

```python
df2['venue'].value_counts().plot(kind='bar',figsize=(10,5))
plt.title('Matches per Venue')
plt.xlabel('Venue')
plt.ylabel('Number of matches')
plt.xticks(rotation=90)
plt.show()
```

## 2.Matches Played by the teams

```python
matches_played=df2['team1'].value_counts().add(df2['team2'].value_counts(),fill_value=0).reset_index()
matches_played.columns=['team','matches_played']
matches_played
```

## 3. Who won the most tosses?

```python
df2['toss_winner'].value_counts().plot(kind='bar',color='yellow')
plt.title('Toss Wins By Team')
plt.show()
```

## 4. Match winners and frequency

```python
df2['match_winner'].value_counts().head(10).plot(kind='bar',color='green')
plt.title('Top Match Winners')
plt.show()
```

## 6. Most frequent "Player of the match"

```python
top_players=df['player_of_the_match'].value_counts().head(10)

top_players.plot(kind='bar',color='pink')
plt.title('Top 10 Player of the match winners')
plt.ylabel('No. of Awards')
plt.xticks(rotation=90)
plt.show()
```

## 7. Top Scorers and Their Highest Scores

```python
top_scorers=df2.groupby('top_scorer')['highscore'].max().sort_values(ascending=False).head(10)

top_scorers.plot(kind='barh',color='red')
plt.title('Top 10 Highest individual Scores')
plt.xlabel('Runs')
plt.gca().invert_yaxis()
plt.show()
           
```

## 8. Toss decision impact on match result

```python
toss_result_df=df2.dropna(subset=['match_winner','toss_winner'])

toss_result_df['toss_win_and_match_win']=np.where(
    toss_result_df['toss_winner']==toss_result_df['match_winner'],
    'Yes','No'
)

sns.countplot(x='toss_win_and_match_win', hue='toss_decision',data=toss_result_df)
plt.title('Did toss winner also win the match')
plt.xlabel('Toss Winner' == 'Match Winner')
plt.ylabel('Match Count')
plt.show()
```

## 9. Highest Match Aggregates (Total runs in both innings)

```python
df2['total_runs'] = df2['first_ings_score'] + df2['second_ings_score']
top_matches = df2.sort_values(by='total_runs', ascending=False).head(10)


print(top_matches[['team1', 'team2', 'first_ings_score', 'second_ings_score', 'total_runs']])
```
